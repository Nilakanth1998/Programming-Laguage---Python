##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.

```python
## Syntax
def function_name(parameter):
    """Docstring"""
    # Function Body
    return expression
```

```python
## Why Function ??
num=24

if num%2==0:
    print("The number is Even")
else:
    print("The number is Odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("The number is Even")
    else:
        print("The number is Odd")
```

```python
## Call the function

even_or_odd(45)
```

```python
## Function with Multiple Parameters

def add(a,b):
    return a+b

result=add(2,5)
print(result)
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} Welcome to Mumbai")

greet("Nilu")
```

```python
## Default Parameters

def greet(name="Guest"):
    print(f"Hello {name} Welcome to Mumbai")

greet()
```

```python
## Variable length Arguments
### Positional & Keywords Arguments

## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"args")
```

```python
## Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Nilu",Age=28,country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f" Positional argument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,5,6,"Nilu",name="Nilu",Age=28,country="India")
```

```python
## Return Statement

def multiply(a,b):
    return a*b

multiply(3,5)
```

```python
## Return multiple parameters

def multiply(a,b):
    return a*b,a

multiply(3,5)
```
